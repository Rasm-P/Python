```python
#Home exercise with request module (not hand-in)
import json
import api_keys
import requests
import pandas as pd
from pandas.io.json import json_normalize

#1,2
url = "http://api.openweathermap.org/data/2.5/forecast"
req = requests.get(url, params={'q': 'Copenhagen,dk', 'mode': 'json', 
                                'units': 'metric', 
                                'appid': api_keys.OWM_API_KEY})
req_json = json.loads(req.text)

#3
def get_main_data():
    forecasts = req_json['list']
    forecasts_tbl = json_normalize(forecasts)
    return forecasts_tbl
get_main_data()

#4
print('Lowest temp: ', max(get_main_data()['main.temp']))
print('Highest temp: ', min(get_main_data()['main.temp']))
```

```python
#City data

#5
file_name = 'city.list.json'

#6
with open(file_name, encoding='utf-8') as f:
    d = json.load(f)
    
#7
def all_available_cities():
    contry_dit = {}
    for c in d:
        if c['country'] not in contry_dit:
            contry_dit[c['country']] = [c['name']]
        else:
            contry_dit[c['country']].append(c['name'])
    return contry_dit
all_available_cities()
```

```python
#8
dk_cities = all_available_cities()['DK']
print(len(dk_cities))
```
