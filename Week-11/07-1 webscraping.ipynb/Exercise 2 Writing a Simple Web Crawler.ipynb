```python
import bs4
import os
import sys
import requests
import shutil

#Write a simple web crawler. More precisely, a program that extracts recursively all links from web pages. The result of 
#running the web crawler is a dictionary, were the key-value pairs correspond to outgoiung links from a web page with the 
#URL, which is stored in the key.

def scrape_links(from_url, max_end, for_depth=1, all_links={}):
    if max_end == for_depth:
        return all_links
    else:
        r = requests.get(from_url)
        r.raise_for_status()
        soup = bs4.BeautifulSoup(r.text, 'html.parser')
        for img in soup.select('a'): 
            if img.get('href') and img.get('href').startswith('http'):
                all_links[len(all_links)] = img.get('href')
                print(all_links)
                all_links = scrape_links(img.get('href'),max_end,for_depth+1,all_links)
                    
scrape_links('https://www.dr.dk/radio/programmer',3)       
```
