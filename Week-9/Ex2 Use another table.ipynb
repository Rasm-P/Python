```python
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt

#Choose any of the other tables in 'databanken' to find interesting data.

#1 Collect the data

files = glob('C:/Users/rasmu/Desktop/FOLK1D.csv')
civilstand_csv = files[0]
civilstand = pd.read_csv(civilstand_csv, sep = ';')
civilstand
```

```python
#2,3,4 Pose 5 or more interesing questions to the data

#-1. How many foreign citizens were female in Aalborg from 2008 to now?

female_list = civilstand[civilstand['KØN'] == 'Kvinder']
female_list_aalborg = female_list[female_list['OMRÅDE'] == 'Aalborg']
female_list_aalborg_foreign = female_list_aalborg[female_list_aalborg['STATSB'] == 'Udenlandsk statsborger']
female_list_aalborg_foreign_all = female_list_aalborg_foreign[female_list_aalborg_foreign['ALDER'] == 'I alt']
female_list_aalborg_foreign_all

female_list_aalborg_foreign_all.plot(x="TID", y="INDHOLD")
```

```python
#-2. How many danish male citizens were 100 years old from 2008 to now?

male_list = civilstand[civilstand['KØN'] == 'Mænd']
male_list_danish = male_list[male_list['STATSB'] == 'Dansk statsborger']
male_list_danish_100 = male_list_danish[male_list_danish['ALDER'] == '100 år']
male_list_danish_100_alt = male_list_danish_100[male_list_danish_100['OMRÅDE'] == 'Hele landet']
male_list_danish_100_alt

male_list_danish_100_alt.plot(x="TID", y="INDHOLD")
```

```python
#-3. What is the destribution over danish males and their age in 2011?

male_list = civilstand[civilstand['KØN'] == 'Mænd']
male_list_danish = male_list[male_list['STATSB'] == 'Dansk statsborger']
male_list_danish_landet = male_list_danish[male_list_danish['OMRÅDE'] == 'Hele landet']
male_list_danish_landet_2011 = male_list_danish_landet[-male_list_danish_landet['TID'].apply(lambda x: (x[0:4] != '2011'))]
male_list_danish_landet_2011_alt = male_list_danish_landet_2011[-male_list_danish_landet_2011['ALDER'].apply(lambda x: (x == 'I alt'))]
male_list_danish_landet_2011_alt

plt.bar(male_list_danish_landet_2011_alt['ALDER'], male_list_danish_landet_2011_alt['INDHOLD'], width=0.5, align='center' , color='blue')
male_list_danish_landet_2011_alt.plot(x="ALDER", y="INDHOLD")
```

```python
#-4. What is the distribution for all ages in denmark in 2020?

all_ages = civilstand[civilstand['OMRÅDE'] == 'Hele landet']
all_ages_stat = all_ages[all_ages['STATSB'] == 'I alt']
all_ages_stat_2020 = all_ages_stat[-all_ages_stat['TID'].apply(lambda x: (x[0:4] != '2020'))]
all_ages_stat_2020_alt = all_ages_stat_2020[-all_ages_stat_2020['ALDER'].apply(lambda x: (x == 'I alt'))]
all_ages_stat_2020_alt_alle = all_ages_stat_2020_alt[all_ages_stat_2020_alt['KØN'] == 'I alt']
all_ages_stat_2020_alt_alle

all_ages_stat_2020_alt_alle.plot(x="ALDER", y="INDHOLD")
```

```python
#-5. What is the diatribution for all persons in Aalborg from 2008 to now?

all_ages = civilstand[civilstand['OMRÅDE'] == 'Aalborg']
all_ages_stat = all_ages[all_ages['STATSB'] == 'I alt']
all_ages_stat_allalder = all_ages_stat[all_ages_stat['ALDER'] == 'I alt']
all_ages_stat_allalder_allekøn = all_ages_stat_allalder[all_ages_stat_allalder['KØN'] == 'I alt']
all_ages_stat_allalder_allekøn

all_ages_stat_allalder_allekøn.plot(x="TID", y="INDHOLD")
```
