```python
#1,2

Article = ["a","an","the","thea"]
Adjective = ["beautiful","sleeveless","meaningless","nicer","adorable","gorgeous","breakable","homeless"]
Noun = ["boy","girl","hands","bridge","town","yarn","house","garden","umbrella","puddle"]
Verb = ["eats","walks","drinks","thinks","runns","smiles","crys"]

for ar in Article:
    for aj in Adjective:
        for n in Noun:
            for v in Verb:
                print(ar+" "+aj+" "+n+" "+v)
```
