```python
%matplotlib notebook
import matplotlib.pyplot as plt

student_attendance = {'day1':33, 'day2':34,'day3':29,'day4':31,'day5':28,'day6':26,'day7':30}
keys = student_attendance.keys()
values = student_attendance.values()

squares = [keys for keys in values]
plt.xlabel("Days", fontsize=10)
plt.ylabel("Students", fontsize=10)
plt.plot(squares)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

people = {'Holger':25,'Helga':54,'Hasse':76,'Halvor':12,'Hassan':43,'Hulda':31,'Hansi':102}
sorted_people = {k: v for k, v in sorted(people.items(), key=lambda item: item[1])}

plt.bar(sorted_people.keys(), sorted_people.values(), width=0.5, align='center')
title = 'Distribution of names and amounts'
plt.title(title, fontsize=12)
plt.xlabel("People", fontsize=10)
plt.ylabel("Amount", fontsize=10)
plt.tick_params(axis='both', which='major', labelsize=10)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import kkdata

def get_population_stat():  
    stats = {}
    years = kkdata.STATISTICS.keys()
    for year_of_interest in years:    
        neighbourhoods = kkdata.STATISTICS[year_of_interest].keys()
        age_range = set([])
        for n in neighbourhoods:
            age_range.update(kkdata.STATISTICS[year_of_interest][n].keys()) #update() on set adds any new elements

        no_citicens_per_age = 0

        for n in neighbourhoods:
            for year in age_range:
                if year in kkdata.STATISTICS[year_of_interest][n].keys():
                    c_codes = set(kkdata.STATISTICS[year_of_interest][n][year].keys())
                    for f_code in c_codes:
                        no_citicens_per_age += kkdata.STATISTICS[year_of_interest][n][year][f_code]
        stats.setdefault(year_of_interest, 0)
        stats[year_of_interest] += no_citicens_per_age
    return stats


stats = get_population_stat()
stat_keys = stats.keys()
stat_values = stats.values()

plt.figure()
plt.xlabel("Years", fontsize=10)
plt.ylabel("People", fontsize=10)
x_values = stat_keys
y_values = stat_values
plt.scatter(x_values, y_values)
```

```python
data = {'apple': 10,'banana':2,'citrus':32,'dragon fruit':8}

explode = (0.1, 0, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(data.values(), labels=data.keys(), explode=explode, autopct=lambda p:'{:.2f}%({:.0f})'.format(p,(p/100)*sum(data.values())), shadow=True, startangle=90)
ax1.set_aspect('equal')
ax1.legend(data.keys(), loc='upper right')
plt.show()
```
