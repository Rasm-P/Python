```python
%matplotlib notebook
import matplotlib.pyplot as plt

people = {'Holger':25,'Helga':54,'Hasse':76,'Halvor':12,'Hassan':43,'Hulda':31,'Hansi':102}
sorted_people = {k: v for k, v in sorted(people.items(), key=lambda item: item[1])}

plt.bar(sorted_people.keys(), sorted_people.values(), width=0.5, align='center')
title = 'Distribution of names and amounts'
plt.title(title, fontsize=12)
plt.xlabel("People", fontsize=10)
plt.ylabel("Amount", fontsize=10)
plt.tick_params(axis='both', which='major', labelsize=10)
```

```python
import kkdata

def get_population_stat(year_of_interest):  
    neighbourhoods = kkdata.STATISTICS[year_of_interest].keys()
    age_range = set([])
    for n in neighbourhoods:
        age_range.update(kkdata.STATISTICS[year_of_interest][n].keys()) #update() on set adds any new elements

    no_citicens_per_age = 0

    for n in neighbourhoods:
        for age in age_range:
            if age in kkdata.STATISTICS[year_of_interest][n].keys():
                c_codes = set(kkdata.STATISTICS[year_of_interest][n][age].keys())
                for f_code in c_codes:
                    no_citicens_per_age += kkdata.STATISTICS[year_of_interest][n][age][f_code]

    return no_citicens_per_age


def add_keys_from_one_another(one, other):
    for k, v in one.items():
        if not k in other.keys():
            other[k] = 0

            
stats = get_population_stat(2015)
years = kkdata.STATISTICS.keys()

squares = [keys for keys in values]
plt.xlabel("Days", fontsize=10)
plt.ylabel("Students", fontsize=10)
plt.plot(squares)

print(years)
print(stats)
```
