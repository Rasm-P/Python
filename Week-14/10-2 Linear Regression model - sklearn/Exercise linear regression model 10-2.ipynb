```python
#Are there a linear relationship here in this csv data:

import pandas as pd
import numpy as np

data = [
['year','GDP(trillion)','4wheeler_car_sale'],
[2011,6.2,26.3],
[2012,6.5,26.65],
[2013,5.48,25.03],
[2014,6.54,26.01],
[2015,7.18,27.9],
[2016,7.93,30.47]
]

#1. save data in a file: car_sales.csv
relation_data = np.array(data)
pd.DataFrame(relation_data).to_csv("car_sales.csv")
```

```python
#2. plot car s3les as a function to GDP (is there a linear relationship?)

car_sales = pd.read_csv('car_sales.csv')
print(car_sales)

car_sales=car_sales.astype(float)

car_sales.plot(x=1,y=2)
car_sales.plot.scatter(x = 1, y = 2)
#To some extent there is a linear relationship
```

```python
#3. fit data to a klearn linear regression model

import sklearn.linear_model

xs = car_sales['GDP(trillion)']
ys = car_sales['4wheeler_car_sale']

xs_reshape = np.array(xs).reshape(-1, 1)

model = sklearn.linear_model.LinearRegression()
model = model.fit(xs_reshape, ys)

print(model.coef_)
print(model.intercept_)
```

```python
#4. predict sales if GDP hits 9 trillion lakhs
predicted = model.predict(xs_reshape)
spending9tril = model.predict([[9.0]])
print(spending9tril)
```
